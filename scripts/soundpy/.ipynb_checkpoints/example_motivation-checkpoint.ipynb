```python
import numpy as np
import pandas as pd
import scipy
from numpy.random import default_rng

rng = default_rng(seed=1)
from sklearn.utils import resample
import matplotlib.pyplot as plt
from scipy.stats import ksone


gridspec_kw = {
    "hspace": 0.00,
    "wspace": 0.00,
    "left": 0.05,
    "right": 0.98,
    "bottom": 0.06,
    "top": 0.98,
}

def plot_p(p, ax=None, figsize=(5, 2.5)):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=figsize, gridspec_kw=gridspec_kw)
    ax.plot(p["t"], p["l"], "D", markersize=3, mfc="black", mec="black")
    ax.set_ylim(-3, 12)
    ax.errorbar(p["t"], p["l"], yerr=p["l_u"], fmt="none", ecolor="black", capsize=0)
    ax.set_xlabel("Time $[$s$]$")
    ax.set_ylabel("Load $[$W$]$")
    ax.grid(
        **{
            "color": "#4f8c9d",
            "alpha": 0.8,
            "linestyle": ":",
            "linewidth": 0.5,
        }
    )
    ax.tick_params(
        direction="in",
        which="both",
        right=True,
        top=True,
        labelsize="x-large",
    )
    # ax.set(**gridspec_kw)

# plot results of check_unary_point_wise as a function of time
def plot_check_unary_point_wise(p, results, ax=None, figsize=(5, 2.5)):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=figsize, gridspec_kw=gridspec_kw)
    ax.plot(p["t"], results, "D", markersize=3, mfc="black", mec="black")
    ax.set_ylim(-0.1, 1.1)
    ax.set_xlabel("Time $[$s$]$")
    ax.set_ylabel("Ratio of satisfaction")
    ax.grid(
        **{
            "color": "#4f8c9d",
            "alpha": 0.8,
            "linestyle": ":",
            "linewidth": 0.5,
        }
    )
    ax.tick_params(
        direction="in",
        which="both",
        right=True,
        top=True,
        labelsize="x-large",
    )
    # ax.set(**gridspec_kw)

def check_unary_point_wise(x, eps, fn_test, n_samples=10):
    """apply a function to every element in the series using bootstrap_single and return an array of the ratio of successes"""
    results = np.zeros(len(x))

    # create samples of x by repeating original values and adding gaussian noise with std=eps
    x_samples = np.zeros((n_samples, len(x)))
    for i in range(n_samples):
        x_samples[i] = x + np.random.normal(0, eps, len(x))

    # apply fn_test to every element in x_samples and store count of trues in results
    for i in range(len(x)):
        results[i] = np.sum(fn_test(x_samples[:, i])) / n_samples

    return results
```

```python
# example with something being larger than zero

p_zero = pd.DataFrame({"t": [1, 2, 4, 8, 9, 10],
                       "l": [6, 5, 3, -0.5, 3, 4],
                       "l_u": [2.1, 0.4, 0.6, 1.5, 2.2, 1.1],
                       })

plot_p(p_zero)
plt.axhline(0, color='grey', linestyle=':')

# point-wise unary constraint that p_zero is larger than zero
results = check_unary_point_wise(p_zero.l, p_zero.l_u,
                                 fn_test= lambda x : x > 0,n_samples=100)

plot_check_unary_point_wise(p_zero, results)
# set the margins of the plot (for the x_label) to something reasonable
```

```python
from scipy import stats
import numpy as np

import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, interact_manual

@interact
def beta_plot(a=5, b=10):
    fig, axs = plt.subplots(1, 2, figsize=(15,5))
    beta = lambda x, a, b: stats.beta.pdf(x, a, b)
    axs[0].plot(np.linspace(0, 1, 100), beta(np.linspace(0, 1, 100), a, b))
    axs[0].set_xlabel(f'f({a},{b})')
    axs[0].set_ylabel('density')
    axs[1].plot(np.linspace(0, 1, 100), beta(np.linspace(0, 1, 100), 1.005, 1.005))
    axs[1].set_xlabel('f(1,1)')
    axs[1].set_ylabel('density')
    
beta_plot()
```
